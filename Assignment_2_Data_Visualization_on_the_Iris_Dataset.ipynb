```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from seaborn import load_dataset
df = load_dataset("iris")
print("First 5 Rows of the Iris Dataset:")
print (df.head())
```

```python
df.info()
print("Missing Values:")
print(df.isnull().sum())
```

```python
print("Statistical Summary:")
print(df.describe())
```

```python
# Question 1- What is the distribution of sepal lengths and widths among the Iris species?
# KDE Plot- To visualize the distribution of sepal length
plt.figure(figsize=(8, 5))
sns.kdeplot(df["sepal_length"], fill=True, color='blue')
plt.title("Kernel Density Estimate of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Density")
plt.show()
```

```python
# Observation- Sepal lengths between 5- 6.5 cm have the highest density
```

```python
# Violin Plot- To visualize the distribution of sepal length
plt.figure(figsize=(8, 5))
sns.violinplot(x=df["species"], y=df["sepal_length"])
plt.title("Violin Plot of Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

```python
# Observation - Sepal lengths of virginica species are highest whereas setosa's lengths are lowest
```

```python
# Histogram - To visualize the distribution of Sepal Width
plt.figure(figsize=(8, 5))
plt.hist(df["sepal_width"], bins=20, color='green', alpha=0.5, edgecolor='black')
plt.title("Histogram of Sepal Width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
#  Observation - Sepal width of 3cm has the highest frequency whereas 4cm has the lowest
```

```python
# Violin Plot- To visualize the distribution of sepal width
plt.figure(figsize=(8, 5))
sns.violinplot(x=df["species"], y=df["sepal_width"])
plt.title("Violin Plot of Sepal Width by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python
# Observation - Setosa has highest sepal width whereas versicolor has lowest sepal width
```

```python
# Question 2 - How does the petal length vary across different Iris species?
# Violin Plot - To visualize the distribution of petal length among different species
plt.figure(figsize=(8, 5))
sns.violinplot(x=df["species"], y=df["petal_length"])
plt.title("Violin Plot of Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.show()
```

```python
# Observation- Virginica has the highest petal lengths and setosa is having lowest petal lengths
```

```python
# Question 3 - Is there a relationship between sepal legth and petal length?
# Scatter Plot - To analyze the relationship between Sepal Length and Petal Length
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["sepal_length"], y=df["petal_length"], hue=df["species"])
plt.title("Scatter Plot of Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.show()
```

```python
# Observations- Virginica is having higher sepal lengths and higher petal lengths whereas Setosa is having lower lengths of both
# Petal lengths are increasing with the increase in sepal lengths
```

```python
# Question 4 - How does the petal width vary across different Iris species?
# Violin Plot - To visualize the distribution of petal width among different species
plt.figure(figsize=(8, 5))
sns.violinplot(x=df["species"], y=df["petal_width"])
plt.title("Violin Plot of Petal Width by Species")
plt.xlabel("Species")
plt.ylabel("Petal Width (cm)")
plt.show()
```

```python
# Observation- Virginica has highest petal width whereas setosa has lowest
```

```python
# Question 5-Are there any patterns observable in the pairwise relationships between features (sepal length, sepal width, petal length, petal width)?
# Scatter Plot - To analyze the relationship between Sepal Length, Sepal Width, Petal Length and Petal Width
plt.figure(figsize=(10, 8))
sns.pairplot(df, hue="species")
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

```python
# Observation - Virginica has highest sepal length, petal length and highest petal width
# Setosa has highest sepal width but lowest petal width and lowest sepal length
```

```python
# Question 6- Can we distinguish between Iris species based on sepal and petal measurements?
# Heatmap to visualize correlation between sepal and petal measurements
plt.figure(figsize=(8, 5))
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Observation- There is least correlation between sepal width-petal length and petal length-sepal width
```
